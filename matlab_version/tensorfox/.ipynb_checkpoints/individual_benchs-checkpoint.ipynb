```python
import numpy as np
from numba import njit, prange
from matplotlib import pyplot as plt
import TensorFox as tfx
import tensorly as tl
from tensorly.decomposition import parafac
import time 
import scipy
```

```python
def swimmer():
    T = np.load('swimmer.npy') 
    r = 50
    return T, r
```

```python
def hw():
    # Hand writing tensor
    T = np.load('handwritten.npy')        
    r = 150        
    return T, r
```

```python
def Mntensor(n):
    Mn = np.zeros((n**2,n**2,n**2))
    for i in range(0,n):
        for j in range(0,n):
            for k in range(0,n):
                Mn[i*n+j,j*n+k,i*n+k] = 1
    r = int(np.ceil( n**(np.log(7)/np.log(2)) ))
    return Mn, r
```

```python
trials = 100
T, r = swimmer()
Tsize = np.linalg.norm(T)
solutions_cpd = []
diffs = np.zeros((trials, trials))

class options:
    display = 0
                
for i in range(trials):
    print(i)
    factors, T_approx, output = tfx.cpd(T, r, options)  
    solutions_cpd.append(factors)
```

```python
for i in range(trials):
    for j in range(trials):
        if i >= j:
            diffs[i,j] = np.nan
        else:
            diffs[i,j] = np.linalg.norm(solutions_cpd[i] - solutions_cpd[j])/np.linalg.norm(solutions_cpd[i])
```

```python
M = diffs[diffs>0]
a = diffs.flatten()
b = np.log10(a[a>0])

plt.figure(figsize=[10,8])
plt.imshow(np.log10(diffs))
plt.colorbar()
plt.show()
print()
[array, bins, patches] = plt.hist(b, 100)
plt.xlabel('Solutions')
plt.ylabel('Quantity')
plt.show()
```

```python
m, n, p = T.shape
I = np.zeros((2, r*(m+n+p)))
I[0,0] = 1
I[1,1] = 1
u = np.random.randn(r*(m+n+p), 1)
u = u/np.linalg.norm(u)
U = np.zeros((2, r*(m+n+p)))
for i in range(2):
    for j in range(r*(m+n+p)):
        U[i,j] = u[i]*u[j]
P = I - 2*U

projected_solutions = np.zeros((len(solutions_cpd), 2))
for i in range(len(solutions_cpd)):
    vector = np.dot(P, solutions_cpd[i])
    projected_solutions[i, :] = vector
```

```python
plt.figure(figsize=[10,6])
plt.plot(projected_solutions[:,0], projected_solutions[:,1], '.')
plt.show()
```
